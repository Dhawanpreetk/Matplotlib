```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
combined_data= pd.merge(mouse_metadata,study_results)

# Display the data table for previe
#combined_data.head(10)
#combined_data.columns

#Extract the columns
combined_df = combined_data[["Mouse ID","Timepoint","Tumor Volume (mm3)","Metastatic Sites","Drug Regimen","Sex","Age_months","Weight (g)"]]
combined_df["Tumor Volume (mm3)"] = combined_df["Tumor Volume (mm3)"].astype(float).map("{:,.2f}".format)

combined_df
```

```python
# Checking the number of mice.
#new_df=combined_df.drop_duplicates()
number_of_mice=combined_df["Mouse ID"].nunique()
number_of_mice
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint duplicated(subset=["Mouse ID","Timepoint"])
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint.
duplicate_val=combined_df[combined_df.duplicated(subset=["Mouse ID","Timepoint"])]["Mouse ID"]

print(duplicate_val[909],"dtype =",duplicate_val.dtype )
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
new_df=combined_df[combined_df['Mouse ID']== "g989" ] #df2 = df[df['Courses'].duplicated() == True]
new_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
Clean_df = combined_df[combined_df['Mouse ID'].isin(duplicate_val)==False]
Clean_df[['Tumor Volume (mm3)']] = Clean_df[['Tumor Volume (mm3)']].apply(pd.to_numeric)
Clean_df
```

```python
# Checking the number of mice in the clean DataFrame.
Clean_df["Mouse ID"].nunique()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
drug_regimen =Clean_df['Drug Regimen'].unique()
#tumor_volume = Clean_df['Tumor Volume (mm3)']
Mean_tumor=Clean_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].mean()
Median_tumor=Clean_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].median()
var_tumor=Clean_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].var()
std_tumor=Clean_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].std()
stderr_tumor=Clean_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].sem()




# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.
 
summary_df = pd.DataFrame({'Mean Tumor Volume':Mean_tumor,'Median Tumor Volume':Median_tumor,'Tumor Volume Variance':var_tumor,'Tumor Volume Std.Dev.':std_tumor,'Tumor Volume Std.Err.':stderr_tumor})    
summary_df
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)
# Using the aggregation method, produce the same summary statistics in a single line

Clean_df.groupby(['Drug Regimen']).agg({"Tumor Volume (mm3)":['mean','median','var','std','sem']})


     
```

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
# Start by getting the last (greatest) timepoint for each mouse
#Capomulin = Clean_df.loc[Clean_df['Drug Regimen']=='Capomulin',:]

last_timepoint =Clean_df.groupby(['Mouse ID'])['Timepoint'].max()
last_timepoint_df = last_timepoint.reset_index()


# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
#Cap_merge = pd.merge(Cap_volume,Clean_df,on=["Mouse ID","Timepoint"],how='left')
#Cap_merge

last_timepoint_merge = last_timepoint_df.merge(Clean_df, on=('Mouse ID','Timepoint'),how='left')
last_timepoint_merge
```

```python
# Put treatments into a list for for loop (and later for plot labels)

treatments =['Capomulin',"Ramicane","Infubinol","Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_list = []

for drug in treatments:
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    final_tumor_vol = last_timepoint_merge.loc[last_timepoint_merge["Drug Regimen"] == drug, 'Tumor Volume (mm3)']
    
    # add subset 
    tumor_vol_list.append(final_tumor_vol)
    
  
    quartiles = final_tumor_vol.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    outliers = final_tumor_vol.loc[(final_tumor_vol < lower_bound) | (final_tumor_vol > upper_bound)]
    print(f"{drug}'s potential outliers: {outliers}")

      
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
plt.boxplot(tumor_vol_list,labels=treatments)
plt.title('Tumor Volume at Selected Mouse')
plt.ylabel('Final Tumor Volume (mm3)')
plt.xlabel('Drug Regimen')
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
correlation = st.pearsonr(average["Weight (g)"],average["Tumor Volume (mm3)"])
print(f"The correlation coefficient is {round(correlation[0],2)}")
```
